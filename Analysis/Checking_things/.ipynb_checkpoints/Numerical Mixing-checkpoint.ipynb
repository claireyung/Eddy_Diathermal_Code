```python
def weighted_time_mean(variable):
    month_length = variable.time.dt.days_in_month
    time_mean = (variable*month_length).sum('time')/(month_length.sum('time'))
    return time_mean
```

```python
# load snapshot data
base_msc = '/scratch/e14/cy8964/access-om2/archive/025deg_jra55_ryf/'
# ---------------------------------------------------------------------------------- #
# load in data set ----------------------------------------------------------------- #
fsnap_095 = xr.open_mfdataset(base_msc+'output095/ocean/ocean_snap.nc',combine='by_coords',chunks={'time': 1},decode_times = True).chunk({'time': 1,'yt_ocean':ych/10,'xt_ocean':xch/10})
fsnap_095 = fsnap_095.isel(time = 11)
print(fsnap_095.time.values)
fsnap_105 = xr.open_mfdataset(base_msc+'output105/ocean/ocean_snap.nc',combine='by_coords',chunks={'time': 1},decode_times = True).chunk({'time': 1,'yt_ocean':ych/10,'xt_ocean':xch/10})
fsnap_105 = fsnap_105.isel(time = 11)
print(fsnap_105.time.values)
```

```python
tbins = fwmass_m.neutralrho_edges.values
# find volume in temp,x,y bins
Vol_in_bin_start = histogram(fsnap_095.temp-273.15, bins=[tbins], dim = ['st_ocean'],weights=fsnap_095.dzt*fgrd.area_t).rename({'temp_bin':'neutral','xt_ocean':'grid_xt_ocean','yt_ocean':'grid_yt_ocean'})
# integrate to get curly V (volume below isotherm)
curly_V_start = Vol_in_bin_start.sum('neutral')-gridwm.cumsum(Vol_in_bin_start,'T',boundary="extend")#*dT
# repeat for end time
Vol_in_bin_end = histogram(fsnap_105.temp-273.15, bins=[tbins], dim = ['st_ocean'],weights=fsnap_105.dzt*fgrd.area_t).rename({'temp_bin':'neutral','xt_ocean':'grid_xt_ocean','yt_ocean':'grid_yt_ocean'})
curly_V_end = Vol_in_bin_end.sum('neutral')-gridwm.cumsum(Vol_in_bin_end,'T',boundary="extend")#*dT

# find heat content in each bin
dTheta_dPhi_H_start = rho0*Cp*histogram(fsnap_095.temp-273.15, bins=[tbins], dim = ['st_ocean'],weights=fsnap_095.dzt*fgrd.area_t*(fsnap_095.temp-273.15)).rename({'temp_bin':'neutral','xt_ocean':'grid_xt_ocean','yt_ocean':'grid_yt_ocean'})
dTheta_dPhi_H_end = rho0*Cp*histogram(fsnap_105.temp-273.15, bins=[tbins], dim = ['st_ocean'],weights=fsnap_105.dzt*fgrd.area_t*(fsnap_105.temp-273.15)).rename({'temp_bin':'neutral','xt_ocean':'grid_xt_ocean','yt_ocean':'grid_yt_ocean'})
# difference/time
dHdt = (dTheta_dPhi_H_end-dTheta_dPhi_H_start)/(10*365*24*3600)/area
dHdt = dHdt.sum('neutral')-gridwm.cumsum(dHdt, 'T',boundary="fill", fill_value=0)#*dT
# same for curly V
dVdt = (curly_V_end-curly_V_start)/(10*365*24*3600)/area
```

Calculate terms based on https://github.com/rmholmes/MOM_diatherm_analysis/blob/master/Heat_Budget_Process_Production.m

```python
# as Ryan defined things
JIy = fwmass_mean.ty_trans_nrho/rho0
JIx = fwmass2_mean.tx_trans_nrho/rho0

JS = fwmass2_mean.mass_pmepr_on_nrho/rho0/area
dift = fwmass_mean.temp_vdiffuse_diff_cbt_on_nrho + fwmass_mean.temp_nonlocal_KPP_on_nrho + \
       fwmass_mean.temp_vdiffuse_sbc_on_nrho + fwmass_mean.sw_heat_on_nrho + \
       fwmass_mean.frazil_on_nrho + fwmass2_mean.temp_eta_smooth_on_nrho + \
       fwmass_mean.sfc_hflux_pme_on_nrho + fwmass2_mean.temp_rivermix_on_nrho
QIy = fwmass2_mean.temp_yflux_adv_on_nrho + fwmass2_mean.temp_yflux_submeso_on_nrho
QIx = fwmass2_mean.temp_xflux_adv_on_nrho + fwmass2_mean.temp_xflux_submeso_on_nrho

# want convergence of heat and volume into cell, so take difference in transport between cells
QI = gridwm.diff(QIy,'y',boundary="fill", fill_value=0)/area + gridwm.diff(QIx,'x',boundary="fill", fill_value=0)/area
JI = gridwm.diff(JIy,'y',boundary="fill", fill_value=0)/area + gridwm.diff(JIx,'x',boundary="fill", fill_value=0)/area

# # # all convergences, so integrate to get flux
JI = (JI.sum('neutral')-gridwm.cumsum(JI, 'T',boundary="fill", fill_value=0))
JS = (JS.sum('neutral')-gridwm.cumsum(JS, 'T',boundary="fill", fill_value=0))
dift = (dift.sum('neutral')-gridwm.cumsum(dift, 'T',boundary="fill", fill_value=0))
QI = (QI.sum('neutral')-gridwm.cumsum(QI, 'T',boundary="fill", fill_value=0))

QI = -QI
JI = -JI
```

```python
Te = fwmass_m.neutralrho_edges
ndif = dHdt - (dVdt - JI - JS)*rho0*Cp*Te - dift - QI;
```

```python
(ndif*area/dy).sum('grid_xt_ocean').plot(x = 'grid_yt_ocean',cmap = 'coolwarm')
```
